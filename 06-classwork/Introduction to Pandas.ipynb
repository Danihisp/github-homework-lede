```python
df.groupby(by='continent') \
    .population \
    .sum() \
    .sort_values() \
    .plot(kind='barh')
```

```python
# It plots from the bottom to the top
# So if you wanted to put the smallest on top???
# ascending=False
df.groupby(by='continent') \
    .population \
    .sum() \
    .sort_values(ascending=False) \
    .plot(kind='barh')
```
